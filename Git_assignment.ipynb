Q2.  **Tasks to Be Performed:**


1. Create a Git working directory with feature1.txt and feature2.txt in the master branch

2. Create 3 branches develop, feature1 and feature2

3. In develop branch create develop.txt, do not stage or commit it

4. Stash this file and check out to feature1 branch

5. Create new.txt file in feature1 branch, stage and commit this file

6. Checkout to develop, unstash this file and commit

7. Please submit all the Git commands used to do the above step

**Answer**

    Step 1: Create a Git working directory and initial files

    mkdir git_assignment_q2
    cd git_assignment_q2
    git init
    touch feature1.txt feature2.txt

    Add some simple content:

    echo "This is feature1 file" > feature1.txt
    echo "This is feature2 file" > feature2.txt

    Stage and commit them to master:

    git add feature1.txt feature2.txt
    git commit -m "Add feature1.txt and feature2.txt in master branch"

    Verification:

    git status
    git log --oneline

    ---

    Step 2: Create 3 branches — develop, feature1, and feature2

    git branch develop
    git branch feature1
    git branch feature2

    Verification:

    git branch

    > Output should show:
    * master, develop, feature1, feature2

    ---

    Step 3: Switch to the develop branch and create a file

    git checkout develop
    touch develop.txt
    echo "This is the develop branch file" > develop.txt

    Do not stage or commit yet.

    Verification:

    git status

    > You should see develop.txt as “Untracked file”.

    ---

    Step 4: Stash the uncommitted file

    git stash push -m "Save develop.txt file"

    Verification:

    git stash list

    > You should see something like:
    stash@{0}: On develop: Save develop.txt file


    Now the working directory is clean.

    ---

    Step 5: Checkout to feature1 branch and create a new file

    git checkout feature1
    touch new.txt
    echo "This is a new file in feature1 branch" > new.txt
    git add new.txt
    git commit -m "Add new.txt in feature1 branch"

    Verification:

    git log --oneline

    ---

    Step 6: Go back to develop branch and apply stash

    git checkout develop
    git stash pop

    Verification:

    git status

    > Now develop.txt should reappear as an untracked file.

    ---

    Step 7: Stage and commit the unstashed file

    git add develop.txt
    git commit -m "Add develop.txt after unstashing"

    Verification:

    git log --oneline

    ---

    Final Folder Structure

    git_assignment_q2/
    ├── feature1.txt
    ├── feature2.txt
    ├── develop.txt      (in develop branch)
    └── new.txt          (in feature1 branch)

    ---

    Summary of All Commands

Q3. **Tasks to Be Performed**:


1. Create a Git working directory, with the following branches:


               Develop

               F1

               f2


2. In the master branch, commit main.txt file

3. Put develop.txt in develop branch, f1.txt and f2.txt in f1 and f2 respectively

4. Push all these branches to GitHub

5. On local delete f2 branch

6. Delete the same branch on GitHub as well

**Answer**

    Step 1: Create a new working directory and initialize Git

    mkdir git_assignment_q3
    cd git_assignment_q3
    git init

    ---

    Step 2: Create and commit main.txt in the master branch

    touch main.txt
    echo "This is the main branch file" > main.txt
    git add main.txt
    git commit -m "Add main.txt in master branch"

    => Verification:

    git log --oneline
    git status

    ---

    Step 3: Create branches — develop, f1, and f2

    git branch develop
    git branch f1
    git branch f2

    => Verification:

    git branch

    > Output should show:
    * master, develop, f1, f2

    ---

    Step 4: Add files to each branch

    a. In develop branch:

    git checkout develop
    touch develop.txt
    echo "This is the develop branch file" > develop.txt
    git add develop.txt
    git commit -m "Add develop.txt in develop branch"

    b. In f1 branch:

    git checkout f1
    touch f1.txt
    echo "This is the f1 branch file" > f1.txt
    git add f1.txt
    git commit -m "Add f1.txt in f1 branch"

    c. In f2 branch:

    git checkout f2
    touch f2.txt
    echo "This is the f2 branch file" > f2.txt
    git add f2.txt
    git commit -m "Add f2.txt in f2 branch"

    => Verification (optional):

    git log --oneline

    ---

    Step 5: Push all branches to GitHub

    First, create a new repo on GitHub (e.g., git_assignment_q3),
    then connect your local repo to it:

    git remote add origin https://github.com/<your-username>/git_assignment_q3.git

    Replace <your-username> with your actual GitHub username.

    Now push all branches one by one:

    git push -u origin master
    git push -u origin develop
    git push -u origin f1
    git push -u origin f2

    > Verification:

    git branch -r

    > You’ll see origin/master, origin/develop, origin/f1, and origin/f2.


    ---

    Step 6: Delete f2 branch locally

    git branch -d f2

    => Verification:

    git branch

    > You should now see only:
    master, develop, and f1

    ---

    Step 7: Delete f2 branch on GitHub

    git push origin --delete f2

    => Verification: You can check on GitHub — the f2 branch will no longer appear in the branch list.

    ---

    - Final Branch Summary

    Branch	File Added	Commit Message

    master	main.txt	Add main.txt in master branch
    develop	develop.txt	Add develop.txt in develop branch
    f1	f1.txt	Add f1.txt in f1 branch
    f2	f2.txt (deleted later)	Add f2.txt in f2 branch

    ---

    Folder Structure

    git_assignment_q3/
    ├── main.txt         (in master)
    ├── develop.txt      (in develop)
    ├── f1.txt           (in f1)
    └── f2.txt           (was in f2 branch, deleted later)

Q4. **Tasks to Be Performed**:

1. Put master.txt on master branch, stage and commit

2. Create 3 branches: public 1, public 2 and private

3. Put public1.txt on public 1 branch, stage and commit

4. Merge public 1 on master branch

5. Merge public 2 on master branch

6. Edit master.txt on private branch, stage and commit

7. Now update branch public 1 and public 2 with new master code in private

8. Also update new master code on master

9. Finally update all the code on the private branch

**Answer**
    Step 1: Create a working directory and initialize Git

    mkdir git_assignment_q4
    cd git_assignment_q4
    git init

    ---

    Step 2: Create and commit master.txt on the master branch

    touch master.txt
    echo "This is the master branch file" > master.txt
    git add master.txt
    git commit -m "Add master.txt in master branch"

    => Verification:

    git log --oneline

    ---

    Step 3: Create three new branches — public1, public2, and private

    git branch "public 1"
    git branch "public 2"
    git branch private

    => Verification:

    git branch

    ---

    Step 4: Add and commit public1.txt in public 1 branch

    git checkout "public 1"
    touch public1.txt
    echo "This is the public1 branch file" > public1.txt
    git add public1.txt
    git commit -m "Add public1.txt in public 1 branch"

    ---

    Step 5: Merge public 1 into master

    git checkout master
    git merge "public 1" -m "Merge public 1 branch into master"

    => Verification:

    git log --oneline --graph --all

    ---

    Step 6: Merge public 2 into master

    git checkout "public 2"
    touch public2.txt
    echo "This is the public2 branch file" > public2.txt
    git add public2.txt
    git commit -m "Add public2.txt in public 2 branch"

    git checkout master
    git merge "public 2" -m "Merge public 2 branch into master"

    => Verification:

    git log --oneline --graph --all

    ---

    Step 7: Edit master.txt on private branch

    git checkout private
    echo "This is the updated master.txt content from private branch" > master.txt
    git add master.txt
    git commit -m "Update master.txt on private branch"

    ---

    Step 8: Update public 1 and public 2 branches with new master code from private

    First, merge the private branch into each public branch to sync them.

    git checkout "public 1"
    git merge private -m "Merge updates from private branch into public 1"

    git checkout "public 2"
    git merge private -m "Merge updates from private branch into public 2"

    => Verification:

    git diff master

    > Both public branches now contain the updated master.txt.

    ---

    Step 9: Update the master branch with the latest private code

    git checkout master
    git merge private -m "Merge private updates into master"

    => Verification:

    git log --oneline --graph --all

    ---

    Step 10: Finally, ensure all branches have the latest code (update all on private)

    Now merge all branches into private so private has everything.

    git checkout private
    git merge master -m "Merge master into private for final sync"
    git merge "public 1" -m "Merge public 1 into private"
    git merge "public 2" -m "Merge public 2 into private"

    => Verification:

    git log --oneline --graph --all

    > The private branch now contains all changes from master, public 1, and public 2.

    ---

    - Final Branch Summary

    Branch	Files	Description

    master	master.txt, public1.txt, public2.txt	Final merged master
    public 1	master.txt, public1.txt, public2.txt	Updated with private changes
    public 2	master.txt, public1.txt, public2.txt	Updated with private changes
    private	master.txt, public1.txt, public2.txt	All updates merged here

    ---

    -> Final Folder Structure

    git_assignment_q4/
    ├── master.txt
    ├── public1.txt
    └── public2.txt

Q5. **Tasks to Be Performed**:

1. Create a Git Flow workflow architecture on Git

2. Create all the required branches

3. starting from e feature branch, push the branch to the master, following the architecture

4. Push an urgent.txt on master using hotfix

**Answer**
    Step 1: Create a working directory and initialize Git

    mkdir git_assignment_q5
    cd git_assignment_q5
    git init

    ---

    Step 2: Create Git Flow branch architecture

    In Git Flow, the standard branch model is:

    master → production-ready code

    develop → integration branch for features

    feature/ → individual new features

    release/ → preparing production version

    hotfix/ → urgent fixes to master


    Let’s create them:

    # We already have master after init
    git branch develop
    git branch feature/login
    git branch release/v1.0
    git branch hotfix/urgent-fix

    => Verification:

    git branch

    > You should see: master, develop, feature/login, release/v1.0, hotfix/urgent-fix

    ---

    Step 3: Add and commit base file on master

    touch main.txt
    echo "Initial master version" > main.txt
    git add main.txt
    git commit -m "Initial commit on master"

    ---

    Step 4: Move to develop branch and make a change

    git checkout develop
    touch develop.txt
    echo "This is develop branch setup" > develop.txt
    git add develop.txt
    git commit -m "Add develop.txt on develop branch"

    ---

    Step 5: Start from the feature branch

    git checkout feature/login
    touch feature.txt
    echo "This is the feature branch work" > feature.txt
    git add feature.txt
    git commit -m "Add feature.txt on feature branch"

    Now, merge it into develop, as per Git Flow:

    git checkout develop
    git merge feature/login -m "Merge feature/login into develop"

    => Verification:

    git log --oneline --graph --all

    ---

    Step 6: Merge develop into release (prepare for release)

    git checkout release/v1.0
    git merge develop -m "Prepare release v1.0 from develop branch"

    ---

    Step 7: Merge release into master (deploy final code)

    git checkout master
    git merge release/v1.0 -m "Release v1.0 merged into master"

    => Verification:

    git log --oneline --graph --all

    ---

    Step 8: Push branches to GitHub

    Create a GitHub repo (e.g., git_assignment_q5) and connect it:

    git remote add origin https://github.com/<your-username>/git_assignment_q5.git

    Then push all branches:

    git push -u origin master
    git push -u origin develop
    git push -u origin feature/login
    git push -u origin release/v1.0
    git push -u origin hotfix/urgent-fix

    ---

    Step 9: Handle an urgent fix using hotfix

    git checkout hotfix/urgent-fix
    touch urgent.txt
    echo "This is an urgent fix for production" > urgent.txt
    git add urgent.txt
    git commit -m "Add urgent.txt via hotfix branch"

    Now merge it into master:

    git checkout master
    git merge hotfix/urgent-fix -m "Merge urgent hotfix into master"

    And merge the fix back into develop to keep it up to date:

    git checkout develop
    git merge hotfix/urgent-fix -m "Sync hotfix changes into develop"

    =>  Verification:

    git log --oneline --graph --all

    ---

    => Final Branch Summary

    Branch	Purpose	File(s)

    master	Final production branch	main.txt, develop.txt, feature.txt, urgent.txt
    develop	Active development branch	develop.txt, feature.txt, urgent.txt
    feature/login	Feature work	feature.txt
    release/v1.0	Pre-production testing	main.txt, develop.txt, feature.txt
    hotfix/urgent-fix	Emergency fix	urgent.txt

    ---

    -> Folder Structure

    git_assignment_q5/
    ├── main.txt
    ├── develop.txt
    ├── feature.txt
    └── urgent.txt

---
